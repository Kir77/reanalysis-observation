```python
from mpl_toolkits.basemap import Basemap, cm
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

data1 = sio.loadmat('/Users/qimindeng/research/Germany/comparecac2.mat');
ec = data1['ec']
eg = data1['eg']
peg = data1['peg']
lc = data1['lc']
lg = data1['lg']
plg = data1['plg']
sc = data1['sc']
sg = data1['sg']
psg = data1['psg']
r1 = data1['r1']
r2 = data1['r2']

# plot 
with np.load('/Users/qimindeng/research/Germany/GermanyT.npz') as data:#open the dataset
#initialize
    lat=data['lat']
    lon=data['lon']
    
data=np.transpose(sg)
lon_0 = (lon[0]+lon[-1])/2
lat_0 = (lat[0]+lat[-1])/2

# create figure and axes instances
fig = plt.figure(figsize=(18,25))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
# create polar stereographic Basemap instance.
m = Basemap(projection='lcc',lon_0=lon_0,lat_0=lat_0,\
            llcrnrlat=lat[0],urcrnrlat=lat[-1],\
            llcrnrlon=lon[0],urcrnrlon=lon[-1],\
            resolution='i')
# draw coastlines, state and country boundaries, edge of map.

plt.subplot(331)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
# draw parallels.
parallels = np.arange(0.,90,2.5)
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
# draw meridians
meridians = np.arange(0.,180.,2.5)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
ny = data.shape[0]; nx = data.shape[1]
lons, lats = m.makegrid(nx, ny) # get lat/lons of ny by nx evenly space grid.
x, y = m(lons, lats) # compute map proj coordinates.
# draw filled contours.
clevs=np.linspace(30,110,81)
cs = m.contourf(x,y,data,clevs,cmap=cm.s3pcpn_l)
# add colorbar.
cbar = m.colorbar(cs,location='bottom',pad="5%")
#cbar.set_label('days')
# add title
plt.title('start of growing season')

plt.subplot(332)
data2=np.transpose(sc)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
cs = m.contourf(x,y,data2,clevs,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('start date predicted by CAC')

plt.subplot(333)
data3=np.transpose(psg)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
cs = m.contourf(x,y,data3,clevs,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('start date predicted by HA')

plt.subplot(334)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
c_2=np.linspace(280,340,61)
cs = m.contourf(x,y,np.transpose(eg),c_2,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('end of growing season')

plt.subplot(335)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
cs = m.contourf(x,y,np.transpose(ec),c_2,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('end date predicted by CAC')

plt.subplot(336)
data3=np.transpose(psg)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
cs = m.contourf(x,y,np.transpose(peg),c_2,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('end date predicted by HA')

plt.subplot(337)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
c_3=np.linspace(160,280,121)
cs = m.contourf(x,y,np.transpose(lg),c_3,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('length of growing season')

plt.subplot(338)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
cs = m.contourf(x,y,np.transpose(lc),c_3,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('length predicted by CAC')

plt.subplot(339)
m.drawcoastlines()
m.drawstates()
m.drawcountries()
m.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
m.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
cs = m.contourf(x,y,np.transpose(plg),c_3,cmap=cm.s3pcpn_l)
cbar = m.colorbar(cs,location='bottom',pad="5%")
plt.title('length predicted by HA')
filename = "comparewithcac.png"
plt.savefig(filename)
plt.show()
 
```
